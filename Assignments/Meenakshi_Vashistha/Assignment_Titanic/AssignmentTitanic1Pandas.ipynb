```python
#Read data and describe it
# drop columns with more than 25% missing data
# drop columns having independent values(which do not affect the survival rate).

import pandas as pd
import numpy as np

a=pd.read_csv("titanic_dataset.csv") # index_col=0 won't create new indexes, will take which is already there in data 
print(a)
print(type(a))
```

```python
a.describe() # includes only numerical values
#a.describe(include='all')  # includes all the values
```

```python
print(a.columns)
print(a.shape)
print(a.shape[0])
print(a.shape[1])
```

```python
# Find columns with missingdata
# Print shape of dataset

#both isnull() and isna() are same
b=a.isna().any() # it will tell whether NAN values are present ?, if NaN will present even at any place in a column it will give True
print(b) 
print(a.columns[b]) # it will tell names of col which have NaN values
print(a.shape)
```

```python
a.count()
```

```python
# drop columns with more than 25% missing data

# a.dropna(thresh=327, axis=1) # i have calculated thresh incorrect here- 0.25*1309=327, it should 0.75*1309
b=a.isna().sum()
# print(b)
print()
per=(b/a.shape[0]) * 100
print(per)
print()
thresh=a.shape[0]*0.75
print(thresh)
a.dropna(thresh=981, axis=1, inplace=True)  # dropped columns which don't have atleast 981(0.75*1309) values
a.head()
```

```python
a.isna().sum()  # after dropping col with missing value >25%, only age ,fare, embarked col remains which are less than 25%but more than 0% which we haven't dropped
```

```python
# drop columns having independent values(which do not affect the survival rate).

a.drop(columns=['Unnamed: 0'], inplace=True)
a
```

```python
# check data types of all columns
# convert fare to numeric
# find columns still having missing/na values and also count of missing data
# fill na with mean for fare and age column column.
# drop na values for embarked column.
# dump the dataframe to a csv file 'titanic_filtered.csv'.
```

```python
# r=a.fare.astype("str")
# print(r, type(r))  here, type will be series so no use, it is not converting to str like this
# a.fare.str.split(sep="$") a.fare is a series object and sep won't work here
# print(a.fare[1]) will print 1st value of col-, $151.55
# print(a.fare[0][0]) $211.34[0]- 211.34

a.fare.str[1:].astype("float")
```

```python
# find columns still having missing/na values and also count of missing data

print(a.columns[a.isna().any()]) # name of col which have missing data
print(a.isna().sum())  # total missing values(col wise)
```

```python
# fill na with mean for fare and age column 
print(a.mean())
```

```python
b=a.fare.str[1:].astype("float")# mean can be calculated only on numeric values, so need to convert $211 to 211
#print(b)
print()
r = a.fillna({"fare":b.mean(), "age":a.age.mean()}, inplace=True) # b.mean()-calculate mean of b values(all fare values)
a
```

```python
# to check whether nan values of fare and age got filled?

print(a.isna().sum()) 
print(a.columns[a.isna().any()])  # before , it was Index(['age', 'fare', 'embarked'], dtype='object'), now age and fare are not there in the list of nan values
```

```python
# drop na values for embarked column.
# a['embarked'] = a['embarked'].replace(np.nan, 0)-- this is about replacing nan value with 0 in embarked col
a.dropna(inplace=True)
a
```

```python
print(a.isna().sum()) 
print()
a.count()
```

```python
# for survived column replace 0 with D and 1 with A

a['survived']=a['survived'].replace(0,'D').replace(1,'A')
a['survived']
```

```python
#find the frequency of different values in survived column
a['survived'].describe() # it will show freq of more occurence, for ex in this case frq of 0-809, so haven't show frq of 1
a['survived'].value_counts()
```

```python
# group by gender and survived and see the counts in each category

a['survived']=a['survived'].replace('D',0).replace('A',1)
g=a.groupby(['gender','survived'])
print(g.groups)
print(g.keys)
g.count()
```

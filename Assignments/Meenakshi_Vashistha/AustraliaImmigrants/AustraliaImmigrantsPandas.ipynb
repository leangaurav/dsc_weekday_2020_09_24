```python
#Drop

# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like 
#     "OdName,AreaName,RegName,DevName,1980 ... 2008"
```

```python
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)

aus.drop(aus.index[aus.Type=="Emigrants"],inplace=True)
aus
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
a=aus.isna().sum() # 224 values are missing for 2009 to 2013
print(aus.shape[0]) # no of rows
per_of_missing_data=(a/aus.shape[0])* 100      # per= sum/no of obs--> here no of obs = no of rows
print(per_of_missing_data) # for columns 2009 to 2013, 100% data is missing
aus.dropna(thresh=(0.5*aus.shape[0]) , axis=1, inplace=True ) #axis=1 since i have to delete col 
aus
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like 
#     "OdName,AreaName,RegName,DevName,1980 ... 2008"

aus.drop(["Type", "Coverage", "AREA", "REG", "DEV"], axis=1, inplace=True)
aus
```

```python
# Rename

# Rename columns :OdName -> Country; RegName -> Region
# Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
# Find Total number of immigrants for all years, and create a new column Total in data Frame
```

```python
# Rename columns :OdName -> Country; RegName -> Region

aus.rename(columns={"OdName":"Country", "RegName":"Region"}, inplace= True)
aus
```

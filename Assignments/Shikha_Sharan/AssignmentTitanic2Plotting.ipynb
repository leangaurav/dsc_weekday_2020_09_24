```python
import pandas as pd
import matplotlib.pyplot as plt
s=pd.read_csv(r'C:\Users\Shikha\Desktop\dsc_weekday_2020_09_24\CODE\titanic_filtered.csv')
```

```python
#plot histogram of ages of people
print(s)
plt.figure(figsize=(10,5))
r=s.age
r1=plt.hist(r,bins=20,width=1)
print(r1)
```

```python
# plot histogram of ages of people
plt.figure(figsize=(15,5))
r= plt.hist(s.age,width = 2,bins=10)
print(r)
```

```python
#plot no of people in each pclass on bar chart
plt.figure(figsize=(10,5))
r1=s.groupby("pclass").size()
print(r1)
plt.bar(r1.index,r1.values,width=0.2)
```

```python
# plot a bar-chart depicting the ratio of males to females
plt.figure(figsize=(10,5))
r1=s.groupby("gender").size()
print(r1)
plt.barh(r1.index,r1.values)
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.
df1=s.groupby(["gender","survived"])
r=df1.size()
r1=r.unstack()
print(r1)
r1.plot(kind="bar")
r2=r.unstack("survived")
print(r2)
r2.plot(kind="bar")
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
df1=s.groupby(["gender","survived"])
r=df1.size()
r2=r.unstack("survived")
print(r2)
r2.plot(kind="bar",stacked=True)
```

```python
# plot stacked bar of survived with different pclass in same bar
df1=s.groupby(["survived","pclass"])
r=df1.size()
print(r)
r1=r.unstack("survived")
print(r1)
r1.plot(kind="bar",stacked=True)
```

```python
df1=s.groupby(["age","survived"])
r=df1.size()
print(r)
r1=r.unstack("survived")
print(r1)
r1.plot(kind="box")
```

```python
grp = s.groupby(["survived"])
print(grp.size())
print(grp.groups.keys())
d = [ grp.get_group(k).age for k in grp.groups.keys()]
plt.boxplot(d)
plt.show()
```

```python
#  - -
plt.subplot(1,2,1)
plt.plot(s1) # alpha => transparency float : [0-1]

plt.subplot(1,2,2)
plt.plot(s2)
plt.show()
```

```python
plt.figure(figsize=(16,6))
# 1  2
# 3  4
plt.subplot(2,2,1)
plt.plot(s1, c= 'r') # alpha => transparency float : [0-1]

plt.subplot(2,2,2)
plt.plot(s2, c='g')

plt.subplot(2,2,4)
plt.plot(s2//2)
plt.plot(s1)

plt.subplot(2,2,3)
plt.plot(s2+s1)
plt.show()
```

```python
s1 = pd.Series(range(10), index=range(10))
s2 = s1//2 + 3

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.fill_between(s1.index, s1.values, alpha=0.4)
plt.fill_between(s2.index, s2.values, alpha=0.4)
plt.yticks(range(0,20,4))


plt.subplot(1,2,2)
plt.fill_between(s1.index, s1.values, alpha=0.4)
plt.fill_between(s2.index, s2.values + s1.values, s1.values, alpha=0.4)
plt.yticks(range(0,20,4))
plt.show()
```
